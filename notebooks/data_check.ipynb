```python
import glob
import sys,os
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline
```

```python
module_root = '..'
sys.path.append(module_root)
from project import settings as s
```

```python
train_folder = os.path.join(s.raw_folder,'train')
```

```python
training_imgs_right = sorted(glob.glob(train_folder + '/*right.jpeg'))
training_imgs_left = sorted(glob.glob(train_folder + '/*left.jpeg'))
```

```python
def get_nrrd(img_path):
    img = sitk.ReadImage(img_path)
    return sitk.GetArrayFromImage(img)
```

```python
def browse_training_imgs():
    n = len(training_imgs_left)
    def view_image(sample):
        fig = plt.figure(figsize=(12, 6));
        plt.subplot(121)
        plt.imshow(get_nrrd(training_imgs_right[sample]));
        plt.subplot(122)
        plt.imshow(get_nrrd(training_imgs_left[sample]));
    interact(view_image, sample=(0,n-1)) 
```

```python
browse_training_imgs();
```
